### 1. What is Data Cleaning?

Data cleaning is the process of identifying and correcting errors, inconsistencies, and inaccuracies in a dataset. It includes removing duplicate records, handling missing values, fixing incorrect formats, and standardizing data to ensure reliable analysis and decision-making.


### 2. Consequences of Unclean Data

- Incorrect insights and misleading conclusions  
- Inaccurate reports and forecasts  
- Poor business decision-making  
- Loss of credibility and trust in data  
- Financial and operational risks  


### 3. Common Steps in Data Cleaning

1. Identify errors and inconsistencies  
2. Remove duplicate records  
3. Handle missing or null values  
4. Standardize data formats  
5. Validate data accuracy  



### 4. Sorting a Dataset by Department and Salary

**Step-by-step approach:**

1. Select the entire dataset including headers  
2. Go to Data → Sort  
3. Sort by Department in A–Z order  
4. Click Add Level  
5. Sort by Salary from Largest to Smallest  
6. Click OK  


### 5. Text Functions Used in Data Cleaning

- TRIM(): Removes extra spaces from text  
- LEFT(): Extracts characters from the left  
- RIGHT(): Extracts characters from the right  
- MID(): Extracts text from the middle  
- CONCAT(): Combines multiple text strings  


### 6. Role of Date Functions

- TODAY(): Returns the current date  
- NOW(): Returns the current date and time  

These functions help track timelines and automate date-based calculations.


### 7–9. Data Validation (Quantity Column)
